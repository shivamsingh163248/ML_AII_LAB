```python
import numpy as np

# Create 1D array
a = np.array([1, 2, 3])
print("1D Array:", a)

# Create 2D array
b = np.array([[1, 2], [3, 4]])
print("2D Array:\n", b)

# Basic operations
print("Sum:", np.sum(a))
print("Mean:", np.mean(a))
print("Square Root of b:\n", np.sqrt(b))
```

```python
import pandas as pd

# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

# Display DataFrame
print("DataFrame:\n", df)

# Access Columns
print("\nNames:", df['Name'])

# Basic Stats
print("\nAverage Age:", df['Age'].mean())
```

```python
import matplotlib.pyplot as plt

# Simple line plot
x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()

# Bar Plot
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]

plt.bar(names, ages, color='skyblue')
plt.title('Age of People')
plt.xlabel('Name')
plt.ylabel('Age')
plt.show()
```
