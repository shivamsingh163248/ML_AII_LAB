```python
import pandas as pd

# Function to identify outliers using IQR method
def identify_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower bound
    upper_bound = Q3 + 1.5 * IQR  # Upper bound
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers

# Example usage
# Replace 'column_name' with the column you want to analyze
column_name = 'Glucose'  # Example column name
outliers = identify_outliers_iqr(df, column_name)

# Print outliers
print(f"Outliers in {column_name} column:")
print(outliers)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Function to plot box plots for outliers
def plot_outliers(data):
    numeric_columns = data.select_dtypes(include=['int64', 'float64']).columns  # Select numeric columns
    plt.figure(figsize=(15, 10))  # Set the figure size

    for i, column in enumerate(numeric_columns, 1):
        plt.subplot(3, 3, i)  # Create a subplot for each column
        sns.boxplot(data[column], color='skyblue')  # Plot the box plot
        plt.title(f'Box Plot for {column}')  # Set the title
        plt.xlabel('')  # Remove x-axis label for better readability

    plt.tight_layout()  # Adjust layout to prevent overlapping
    plt.show()  # Display the plots

# Call the function to plot outliers
plot_outliers(df)
```
