```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
import tensorflow as tf

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

# Select Birds (class 2) and Planes (class 0)
train_filter = np.where((y_train == 0) | (y_train == 2))
test_filter = np.where((y_test == 0) | (y_test == 2))

x_train, y_train = x_train[train_filter], y_train[train_filter]
x_test, y_test = x_test[test_filter], y_test[test_filter]

# Preprocess data
x_train = x_train.reshape(len(x_train), -1) / 255.0
x_test = x_test.reshape(len(x_test), -1) / 255.0

# Map labels to binary
y_train = (y_train == 2).astype(int)
y_test = (y_test == 2).astype(int)

# Train logistic regression model
model = LogisticRegression(max_iter=100, verbose=1)
model.fit(x_train, y_train.ravel())

# Predictions
y_pred = model.predict(x_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from PIL import Image as PILImage

# Step 1: Data Loading & Exploration (No Extraction)
# Load labels
labels_df = pd.read_csv('/content/trainLabels.csv')
print("Sample of Train Labels:")
print(labels_df.head())

# Check the distribution of classes
class_distribution = labels_df['label'].value_counts()
print("\nClass Distribution:")
print(class_distribution)

# Display sample images from each class
def display_sample_images(labels_df, folder_path, samples_per_class=2):
    unique_labels = labels_df['label'].unique()
    plt.figure(figsize=(10, 10))

    for idx, label in enumerate(unique_labels):
        sample_paths = labels_df[labels_df['label'] == label]['id'].values[:samples_per_class]
        for i, img_id in enumerate(sample_paths):
            img_path = os.path.join(folder_path, f"{img_id}.png")
            image = PILImage.open(img_path)
            plt.subplot(len(unique_labels), samples_per_class, idx * samples_per_class + i + 1)
            plt.imshow(image)
            plt.title(label)
            plt.axis('off')

    plt.show()

display_sample_images(labels_df, "/content/train/train", samples_per_class=2)
```

```python
import numpy as np
import pandas as pd
import os
from PIL import Image as PILImage

# Step 2: Data Preprocessing
labels_df = pd.read_csv('/content/trainLabels.csv')

# Filter dataset for Birds and Airplanes
filtered_df = labels_df[labels_df['label'].isin(['airplane', 'bird'])]
print("Filtered Data:")
print(filtered_df.head())

# Load images and labels
train_images = []
train_labels = []
folder_path = "/content/train/train"

for index, row in filtered_df.iterrows():
    img_path = os.path.join(folder_path, f"{row['id']}.png")
    image = PILImage.open(img_path)
    image_array = np.array(image)
    train_images.append(image_array)
    train_labels.append(0 if row['label'] == 'airplane' else 1)  # 0: Airplane, 1: Bird

train_images = np.array(train_images)
train_labels = np.array(train_labels)

# Normalize images (0-1 range)
train_images = train_images / 255.0

# Reshape images to (num_samples, height * width * channels)
train_images = train_images.reshape(len(train_images), -1)

print("Preprocessing Complete: Data Ready for Modeling.")
print("Shape of Train Images:", train_images.shape)
print("Shape of Train Labels:", train_labels.shape)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Step 3: Data Partitioning (Train/Test Split)
# Assuming train_images and train_labels are already loaded from Step 2

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    train_images, train_labels,
    test_size=0.2,
    stratify=train_labels,
    random_state=42
)

print("Training Set Size:", X_train.shape)
print("Testing Set Size:", X_test.shape)
print("Training Labels Distribution:", np.bincount(y_train))
print("Testing Labels Distribution:", np.bincount(y_test))

print("Data Partitioning Complete!")
```

```python
import numpy as np
from sklearn.decomposition import PCA

# Step 4: Dimensionality Reduction (PCA)
# Assuming X_train and X_test are loaded from Step 3

# Applying PCA to reduce dimensions
pca = PCA(n_components=100)  # Retaining 100 principal components
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

print("Original Shape:", X_train.shape)
print("Reduced Shape (PCA):", X_train_pca.shape)

explained_variance = np.sum(pca.explained_variance_ratio_) * 100
print(f"Explained Variance by 100 Components: {explained_variance:.2f}%")

print("Dimensionality Reduction Complete!")
```

```python
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Step 5: SVM Implementation & Classification
# Assuming X_train_pca, X_test_pca, y_train, y_test are loaded from Step 4

# Train SVM Classifier
svm_classifier = SVC(kernel='linear', C=1, random_state=42)
svm_classifier.fit(X_train_pca, y_train)

# Predictions
y_pred = svm_classifier.predict(X_test_pca)

# Evaluation
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=["Airplane", "Bird"]))

# Confusion Matrix Visualization
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Airplane", "Bird"], yticklabels=["Airplane", "Bird"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

print("SVM Classification Complete!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Step 6: Visualization & Evaluation
# Assuming X_train_pca, X_test_pca, y_train, y_test, svm_classifier are loaded from Step 5

# Visualizing Decision Boundaries (Using first 2 principal components)
def plot_decision_boundaries(X, y, model, title):
    h = 0.02  # Step size in mesh
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, alpha=0.8, cmap=ListedColormap(['lightblue', 'lightcoral']))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['blue', 'red']), edgecolors='k')
    plt.title(title)
    plt.xlabel('PCA Component 1')
    plt.ylabel('PCA Component 2')
    plt.show()

# Plot using only the first 2 principal components
svm_2d = SVC(kernel='linear', C=1, random_state=42)
svm_2d.fit(X_train_pca[:, :2], y_train)

plot_decision_boundaries(X_train_pca[:, :2], y_train, svm_2d, "Decision Boundary (Training Data)")
plot_decision_boundaries(X_test_pca[:, :2], y_test, svm_2d, "Decision Boundary (Testing Data)")

print("Visualization Complete!")
```
